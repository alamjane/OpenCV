```python
# Morphological Operators

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_COMPLEX
    cv2.putText(blank_img, text='ABCDE',org=(50,300),fontFace=font,fontScale=5,color=(255,255,255),thickness=10, lineType=cv2.LINE_AA)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
    
    
```

```python
img = load_img()
display_img(img)
```

```python
kernal = np.ones(shape=(5,5), dtype=np.uint8)
```

```python
kernal
```

```python
result = cv2.erode(img, kernal,iterations = 2)
```

```python
display_img(result)
```

```python
black_noise_img = black_noise + img
```

```python
black_noise_img[black_noise_img == -255] = 0
```

```python
black_noise_img.min()
```

```python
closing = cv2.morphologyEx(black_noise_img, cv2.MORPH_CLOSE, kernal)
```

```python
display_img(closing)
```
