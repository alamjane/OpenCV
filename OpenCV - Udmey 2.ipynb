```python
# Blending and Pasting images

import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('C:/Data Science/pic.jpe')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

img2 = cv2.imread('C:/Data Science/cat.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
# Blending images as same size

img1 = cv2.resize(img1,(1200,1200))
img2 = cv2.resize(img2,(1200,1200))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
blended = cv2.addWeighted(src1=img1, alpha=0.8, src2=img2, beta=0.5, gamma=0) 
```

```python
plt.imshow(blended)
```
