## EDGE Detection

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
face = cv2.imread('C:/Data Science/tiger_face.jpeg')
face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)

plt.imshow(face)
```

```python
edges = cv2.Canny(image=face, threshold1 = 127, threshold2 = 127)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=face, threshold1 = 0, threshold2 = 255)
```

```python
plt.imshow(edges)
```

```python
med_val = np.median(face)
med_val
```

```python
# LOWER THRESHOLD TO EITHER 0 OR 70% OF THE MEDIAN VALUE WHICH EVER GREATER
lower = int(max(0,0.7*med_val))

# UPPER THRESHOLD TO EITHER 130% OF THE MEDIAN OR THE MAX 255, WHICH EVER SMALLER
upper = int(min(255,1.3*med_val))
```

```python
edges = cv2.Canny(image=face, threshold1 = lower, threshold2 = upper)
plt.imshow(edges)
```
